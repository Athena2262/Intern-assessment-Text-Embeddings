## ***Embedding to CSV File Format***


**Importing Libraries**

```python
import os
import pandas as pd
import numpy as np
from transformers import BertModel, BertTokenizer
import torch
```

**Load the CSV file**

```python
file_path = '/content/d1.csv'
with open(file_path, 'r') as file:
    lines = file.readlines()
```

**Concatenation**

```python
text_data = ' '.join(lines)
```

**Tokenize the data**

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
tokenized_data = tokenizer.tokenize(text_data)
```

**Splitting the tokens and Initialising empty tensor to store the embeddings**

```python
max_sequence_length = 512
chunks = [tokenized_data[i:i + max_sequence_length] for i in range(0, len(tokenized_data), max_sequence_length)]
all_embeddings = torch.empty(0, 768)
```

**BERT - Pretrained model**

```python
model = BertModel.from_pretrained('bert-base-uncased')
```

**Concatenate the embeddings**

```python
with torch.no_grad():
    for chunk in chunks:
        input_ids = torch.tensor([tokenizer.convert_tokens_to_ids(chunk)])
        outputs = model(input_ids)
        chunk_embeddings = outputs.last_hidden_state.mean(dim=1)
        all_embeddings = torch.cat((all_embeddings, chunk_embeddings), dim=0)

print("Embeddings for the entire sequence:", all_embeddings)
```
